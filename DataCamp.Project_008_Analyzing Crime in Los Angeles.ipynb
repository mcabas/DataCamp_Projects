```python
## 1. Which hour has the highest frequency of crimes? Store as an integer variable called peak_crime_hour.

crimes.describe()

# Ensure 'TIME OCC' is a string and pad with zeros if needed
crimes['TIME OCC str'] = crimes['TIME OCC'].astype(str).str.zfill(4)
crimes.info()
# Extract the hour
crimes['hour'] = crimes['TIME OCC str'].str[:2].astype(int)

# Find the hour with the highest frequency
peak_crime_hour = crimes['hour'].mode()[0]

print(peak_crime_hour)
```

```python
## 2. Which area has the largest frequency of night crimes (crimes committed between 10pm and 3:59am)? Save as a string variable called peak_night_crime_location.

crimes['TIME OCC'] = crimes['TIME OCC'].astype(int)

crimes['night_crimes'] = crimes['TIME OCC'].apply(
    lambda x: 1 if x >= x > 2200 or x < 359 else 0
)

night_crimes = crimes[crimes["night_crimes"] == 1]

print(night_crimes[['Crm Cd Desc', 'TIME OCC', 'night_crimes']].head(10)) 

night_crimes_area =  night_crimes.groupby("AREA NAME").agg(
    # "num_schools" - the number of schools in the borough.
    num_crimes = ("DR_NO", "count")
)

night_crimes_area_index = night_crimes_area.reset_index()

print(night_crimes_area_index)

peak_night_crime_area = night_crimes_area_index[night_crimes_area_index["num_crimes"] == night_crimes_area_index["num_crimes"].max()]

peak_night_crime_location = peak_night_crime_area["AREA NAME"].iloc[0]

print(peak_night_crime_location)
```

```python
## 3. Identify the number of crimes committed against victims of different age groups. Save as a pandas Series called victim_ages, with age group labels "0-17", "18-25", "26-34", "35-44", "45-54", "55-64", and "65+" as the index and the frequency of crimes as the values.

crimes['Vict Age'] = crimes['Vict Age'].astype(int)

# Define the mapping logic
def map_age_groups(VictAge):
    if pd.isna(VictAge):  
        return "N/A"
    elif VictAge <= 17:  
        return "0-17"
    elif VictAge > 17 and VictAge <= 25:
        return "18-25"
    elif VictAge > 25 and VictAge <= 34:
        return "26-34"
    elif VictAge > 34 and VictAge <= 44:
        return "35-44"
    elif VictAge > 44 and VictAge <= 54:
        return "45-54"
    elif VictAge > 54 and VictAge <= 64:
        return "55-64"
    else:
        return "65+"

crimes["age_groups"] = crimes["Vict Age"].apply(map_age_groups)

print(crimes[["Crm Cd Desc", "Vict Age", "age_groups"]].head())

victim_ages = crimes.value_counts("age_groups") 

print(victim_ages)
```
